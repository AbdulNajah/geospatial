## Q1

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(-100,100,100)
y = x**3
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the function
plt.plot(x,y, 'g')

# show the plot
plt.show()
```

### Q4

```python
# hist
plt.figure(figsize = (10,5))
plt.hist(df['Geography'], bins =3, ec= 'k')
plt.xlabel('Country')
plt.ylabel('Count of Customers')
plt.show()

```

```python
# pie-chart

country_df = df['Geography'].value_counts().reset_index().rename(columns = {'index':'Country', 'Geography': 'Count'})
plt.pie(country_df['Count'], labels = country_df['Country'])
plt.show()
```
